```python
import codecs
f = codecs.open( "질문목록 예시.txt", "r", "utf-8" )

#f = open("질문목록 예시.txt", 'r')

lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
# N-gram 으로 분류

def ngram(s, num):
    res = []
    slen = len(s) - num + 1
    for i in range(slen):
        ss = s[i:i+num]
        res.append(ss)
    return res

def diff_ngram(sa, sb, num):
    a = ngram(sa, num)
    b = ngram(sb, num)
    r = []
    cnt = 0
    for i in a:
        for j in b:
            if i == j:
                cnt += 1
                r.append(i)
    return cnt / len(a)


Question = '홍운반점 시키면 몇 분 걸려요?'

# 가장 높은 유사도를 가지는 질문의 index 저장용
highest_index = -1 
highest_value = -1

# 모든 질문 목록에서 비교하기
for i in range(len(lines)):
    value = diff_ngram(Question, lines[i], 2)
    #print(value, " : ",lines[i])
    if value > highest_value:
        highest_index = i
        highest_value = value

print('Highest_accurate_Question: ',lines[highest_index], ' index: ', highest_index, 
      ' Accuracy: ', highest_value)
```

```python
# 레벤슈타인 거리 (편집거리 방법)

def calc_distance(a, b):
    if a == b:
        return 0
    a_len = len(a)
    b_len = len(b)
    if a == "":
        return b_len
    if b == "":
        return a_len
    # 2차원 표 (a_len+1, b_len+2) 준비하기
    matrix = [[] for i in range(a_len+1)]
    for i in range(a_len+1):
        matrix[i] = [0 for j in range(b_len+1)]
    # 0일 때 초기값을 설정
    for i in range(a_len+1):
        matrix[i][0] = i
    for j in range(b_len+1):
        matrix[0][j] = j
    # 표 채우기
    for i in range(1, a_len+1):
        ac = a[i-1]
        for j in range(1, b_len+1):
            bc = b[j-1]
            cost = 0 if (ac == bc) else 1
            matrix[i][j] = min([
                matrix[i-1][j] + 1,     # 문자 삽입
                matrix[i][j-1] + 1,     # 문자 제거
                matrix[i-1][j-1] + cost # 문자 변경
            ])
    return matrix[a_len][b_len]


# result : 2

samples = lines
base = '세 마리 하려면 몇 분 걸려요?\n'

# samples에 저장된 배열을 거리 순서대로 정렬해서 r에 저장
r = sorted(samples, key = lambda n: calc_distance(base, n))
for n in r:
    print(calc_distance(base, n), n)
```
